```python
import kagglehub
import pandas as pd
from pathlib import Path
```

```python
path = kagglehub.dataset_download("omniamahmoudsaeed/real-estate-sales-2001-2022")
```

```python
path
```

```python
data_dir = Path(path)
print(list(data_dir.glob("*")))  # This will show you the available files

csv_file = next(data_dir.glob("*.csv"))  # Picks the first CSV file found
df = pd.read_csv(csv_file)
```

```python
# calculate number of missing
df.isnull().sum()
```

```python
df.isnull().sum() / df.shape[0] * 100
```

```python
df.shape
```

```python
# cleaning the "low" missing values
df = df.dropna(subset=['Date Recorded', 'Address'])
```

```python
df.isnull().sum() / df.shape[0] * 100
```

```python
# setting missing values for Residential type or property type
df['Property Type'] = df['Property Type'].fillna('Unknown')
df['Residential Type'] = df['Residential Type'].fillna('Unknown')
```

```python
df.isnull().sum() / df.shape[0] * 100
```

```python
# dropping the columns which as missing rows
df = df.drop(columns=['Non Use Code', 'Assessor Remarks', 'OPM remarks', 'Location'])
```

```python
df.isnull().sum() / df.shape[0] * 100
```

```python
df['Residential Type'].value_counts()
```

```python
df['Property Type'].value_counts()
```

```python
df['Property Type'].value_counts() / df.shape[0] * 100
```

```python
rare_threshold = 5000
counts = df['Property Type'].value_counts()
rare_types = counts[counts < rare_threshold].index
df['Property Type'] = df['Property Type'].replace(rare_types, 'Other')
```

```python
df['Property Type'].value_counts() / df.shape[0] * 100
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_column = ['Residential Type']
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded_array = encoder.fit_transform(df[['Residential Type']])
```

```python
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_column))
```

```python
df_encoded = pd.concat([df.reset_index(drop=True), encoded_df], axis=1)
```
